```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading data

df = pd.read_csv("https://cocl.us/datascience_survey_data")
df.columns.tolist()
df.index.tolist()
df.shape
df.head()
```

```python
df2 = round((df.mul(100)).div(2233),2)
df=df2
df.sort_values(by=['Very interested'],ascending=False, inplace=True)
df
```

```python
#setting bar colors
color=('#5cb85c', '#5bc0de', '#d9534f')


#ploting bar chart
ax = df.plot(kind='bar', figsize=(20,8), width=0.8, color=color, fontsize=14)


# add title to the plot
plt.title("Percentage of Respondents' Interest in Data Science Areas", fontsize=16)


#removing y axes label
y_axis = ax.axes.get_yaxis()
y_axis.set_visible(False)


#SETTING LEGEND FONT SIZE
ax.legend(fontsize=14)

   


#frame settings
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('grey')
ax.spines['bottom'].set_linewidth(2)


#placing text over bar
def add_value_labels(ax, spacing=5):
    """Add labels to the end of each bar in a bar chart.

    Arguments:
        ax (matplotlib.axes.Axes): The matplotlib object containing the axes
            of the plot to annotate.
        spacing (int): The distance between the labels and the bars.
    """

    # For each bar: Place a label
    for rect in ax.patches:
        # Get X and Y placement of label from rect.
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2

        # Number of points between bar and label. Change to your liking.
        space = spacing
        # Vertical alignment for positive values
        va = 'bottom'

        # If value of bar is negative: Place label below bar
        if y_value < 0:
            # Invert space to place label below
            space *= -1
            # Vertically align label at top
            va = 'top'

        # Use Y value as label and format number with one decimal place
        label = "{:.1f}".format(y_value)

        # Create annotation
        ax.annotate(
            label,                                        # Use `label` as label
            (x_value, y_value),                           # Place label at end of the bar
            xytext=(0, space),                            # Vertically shift label by `space`
            textcoords="offset points",                   # Interpret `xytext` as offset in points
            ha='center',                                  # Horizontally center label
            bbox=dict(boxstyle="round,pad=0.3", fc="gray", alpha=0.4),         # Creating Box around label
            va=va)                                        # Vertically align label differently for positive and negative values.

    
            
# Call the function above.
add_value_labels(ax)


#clear the text of x label
plt.xlabel("")

plt.show()
```
